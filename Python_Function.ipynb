##### Here, I wrote a simple python function that would allow me to quickly return an array of mean values at any location. First, I grouped the dataset of mean values by year, and then used this grouped array to create an equation that selects specific latitude (x) and longitude (y). Finally, based on the inputted x and y values, the python function would return an array of mean values by year for that location

```python
def mean_lat_lon_year(x,y):
    dg = ds.groupby('time.year').mean()
    mean = dg['tp'].sel(latitude=x,longitude=y)

    return mean
```

##### Example run of the python function with x=42.5, y=-71

```python
mean_lat_lon_year(42.5,-71)
```

##### I also created a function that runs on the same code, but with year replaced by month to return monthly mean values.

```python
def mean_lat_lon_month(x,y):
    dm = ds.groupby('time.month').mean()
    mean = dm['tp'].sel(latitude=x,longitude=y)

    return mean
```

##### Example run with x=42.5, y=-71

```python
mean_lat_lon_month(42.5,-71)
```
