```python
import xarray as xr
import matplotlib.pyplot as plt
import numpy as np

import cartopy.crs as ccrs
import cartopy.mpl.ticker as cticker
from cartopy.util import add_cyclic_point
import cartopy.feature as feature
from cartopy.io.shapereader import Reader
from cartopy.feature import ShapelyFeature
```

##### Importing my dataset

```python
file='/home/jstiles4/clim680_dataset/ERA5_Reanalysis_Precip_dataset.grib'
ds=xr.open_dataset(file)
ds
```

##### Converting from m/day to mm/day by multiplying the tp values by 1000

```python
ds = ds*1000
```

##### For this climatology plot, I am interested in plotting the average daily precipitation (tp) by month for the Northeast U.S. I began this by groupby to group the mean tp values by month.

```python
dg_tp=ds.groupby('time.month').mean()
dg_tp
```

##### Here, I composed a series of contour plots for each month on the scale of mean tp. The full plot was organized on a 2 row, 6 column scale, and each individual plot was titled based on which month's data was displayed.

```python
fig = plt.figure(figsize=(18,6))
plt.subplot(2,6,1)
plt.contourf((dg_tp['tp']).sel(month=1))
plt.title('January')
plt.subplot(2,6,2)
plt.contourf((dg_tp['tp']).sel(month=2))
plt.title('February')
plt.subplot(2,6,3)
plt.contourf((dg_tp['tp']).sel(month=3))
plt.title('March')
plt.subplot(2,6,4)
plt.contourf((dg_tp['tp']).sel(month=4))
plt.title('April')
plt.subplot(2,6,5)
plt.contourf((dg_tp['tp']).sel(month=5))
plt.title('May')
plt.subplot(2,6,6)
plt.contourf((dg_tp['tp']).sel(month=6))
plt.title('June')
plt.subplot(2,6,7)
plt.contourf((dg_tp['tp']).sel(month=7))
plt.title('July')
plt.subplot(2,6,8)
plt.contourf((dg_tp['tp']).sel(month=8))
plt.title('August')
plt.subplot(2,6,9)
plt.contourf((dg_tp['tp']).sel(month=9))
plt.title('September')
plt.subplot(2,6,10)
plt.contourf((dg_tp['tp']).sel(month=10))
plt.title('October')
plt.subplot(2,6,11)
plt.contourf((dg_tp['tp']).sel(month=11))
plt.title('November')
plt.subplot(2,6,12)
plt.contourf((dg_tp['tp']).sel(month=12))
plt.title('December')
plt.colorbar()
plt.suptitle('Precipitation (mm/day)')
```

##### This is where I put together full climatology plot, beginning with specifying the number of rows (2) and columns (6). For the projection, I used PlateCarree with a central longitude of 74W (-74). In order to create the climatology plot, I had to run a 'for i' scenario to specify the plot situation for each month in my dataset. I used the contour range of 2.0-4.5 mm/day with intervals of 0.5 mm/day, and assigned the climatology the color scheme of 'YlGnBu.' To make the latitude and longitude ranges show up on the plots, I used the set_xticks and set_yticks functions to specify the ranges of the contour plots. I then assigned each plot a title based on its position (i) within the month dimension, meaning an i value of 0 would represent January with each subsequent i value representing the subsequent months. Finally, I imported all of the necessary Cartopy features, starting with the coastlines, and the state and country borders. I followed them by importing the ocean and lake features with a black color to mask out the data over these ranges, and then imported the Canada shape file with a white color to mask out any data within Canada.

```python
fig, axs = plt.subplots(nrows=2,ncols=6,
                        subplot_kw={'projection': ccrs.PlateCarree(central_longitude=-74)},
                        figsize=(22,8))

axs=axs.flatten()

for i,tp in enumerate(dg_tp['month']):

        data=dg_tp['tp'].sel(month=tp)

        data,lons=add_cyclic_point(data,coord=dg_tp['longitude'])

        cs=axs[i].contourf(lons,dg_tp['latitude'],data, levels=[2.0, 2.5, 3.0, 3.5, 4.0, 4.5, 5.0],
                          transform = ccrs.PlateCarree(),
                          cmap='YlGnBu',extend='both')

        axs[i].set_xticks([-80,-76,-72,-68], crs=ccrs.PlateCarree())
        lon_formatter = cticker.LongitudeFormatter()
        axs[i].xaxis.set_major_formatter(lon_formatter)
        
        axs[i].set_yticks([38,40,42,44,46,48], crs=ccrs.PlateCarree())
        lat_formatter = cticker.LatitudeFormatter()
        axs[i].yaxis.set_major_formatter(lat_formatter)

        axs[0].set_title('January',fontsize=15)
        axs[1].set_title('February',fontsize=15)
        axs[2].set_title('March',fontsize=15)
        axs[3].set_title('April',fontsize=15)
        axs[4].set_title('May',fontsize=15)
        axs[5].set_title('June',fontsize=15)
        axs[6].set_title('July',fontsize=15)
        axs[7].set_title('August',fontsize=15)
        axs[8].set_title('September',fontsize=15)
        axs[9].set_title('October',fontsize=15)
        axs[10].set_title('November',fontsize=15)
        axs[11].set_title('December',fontsize=15)

        axs[i].coastlines()
        
        axs[i].add_feature(feature.OCEAN, zorder=2, color='#1F2A38')
        axs[i].add_feature(feature.BORDERS, zorder=2, color='k')
        axs[i].add_feature(feature.LAKES, zorder=3, color='#1F2A38')
        axs[i].add_feature(feature.BORDERS.with_scale('10m'),
               linestyle='-', alpha=.5)
        axs[i].add_feature(feature.STATES.with_scale('10m'),
               linestyle='-', alpha=.5)
        
        fname='/home/jstiles4/gadm41_CAN_1.shp'
        shape_feature = ShapelyFeature(Reader(fname).geometries(),
                               ccrs.PlateCarree(), edgecolor='black', color='white')
        
        axs[i].add_feature(shape_feature)
        
fig.subplots_adjust(bottom=0.25, top=0.9, left=0.05, right=0.95,
                    wspace=0.2, hspace=0.25)

cbar_ax = fig.add_axes([0.2, 0.175, 0.6, 0.02])

cbar=fig.colorbar(cs, cax=cbar_ax,orientation='horizontal')
cbar.ax.tick_params(labelsize=11)
cbar.set_label(label='mm/day', size='11')

plt.suptitle('Average Daily Precipitation by Month Northeast U.S. 1990-2010',fontsize=15)
```
